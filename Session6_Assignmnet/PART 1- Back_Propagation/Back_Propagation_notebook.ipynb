## Setting the Inputs and Target

```python
i1 = 0.05
i2 = 0.1

t1 = 0.5
t2 = 0.5
```

## Defining the initial set of weights and the hidden layer weights as shown in the image above

```python
w1 = 0.15
w2 = 0.2
w3 = 0.25
w4 = 0.3
w5 = 0.4
w6 = 0.45
w7 = 0.5
w8 = 0.55
```

## To test for multiple Learning Rate for a range of epoch, how the errors are getting populated

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
epochs = 100
lrlist = [0.1,0.2,0.5,0.8,1.0,2.0]

epoch = []
error  = []
lr = []
for j in lrlist:
    Error_total = []
    epochlist = []

    for i in range(epochs):
        epochlist.append(i)
        h1 = i1*w1 + i2*w2 
        a_h1 = 1/(1+math.exp(-1*h1))
        h2 = i1*w3 + i2*w4
        a_h2 = 1/(1+math.exp(-1*h2))
        o1 = w5*a_h1 + w6*a_h2
        a_o1 = 1/(1+math.exp(-1*o1))
        o2 = w7*a_h1 + w8*a_h2
        a_o2 = 1/(1+math.exp(-1*o2))

        E1 = 0.5 * (t1 - a_o1)**2
        E2 = 0.5 * (t2 - a_o2)**2
        E_tot = E1 + E2
        Error_total.append(E_tot)

        dw1 = ((a_o1 - t1) * a_o1 * (1 - a_o1) * w5 +  (a_o2 - t2) * a_o2 * (1 - a_o2) * w7) * a_h1 * (1 - a_h1) * i1
        dw2 = ((a_o1 - t1) * a_o1 * (1 - a_o1) * w5 +  (a_o2 - t2) * a_o2 * (1 - a_o2) * w7) * a_h1 * (1 - a_h1) * i2
        dw3 = ((a_o1 - t1) * a_o1 * (1 - a_o1) * w6 +  (a_o2 - t2) * a_o2 * (1 - a_o2) * w8) * a_h2 * (1 - a_h2) * i1
        dw4 = ((a_o1 - t1) * a_o1 * (1 - a_o1) * w6 +  (a_o2 - t2) * a_o2 * (1 - a_o2) * w8) * a_h2 * (1 - a_h2) * i2
        dw5 = (a_o1 - t1) * a_o1 * (1 - a_o1) *  a_h1
        dw6 = (a_o1 - t1) * a_o1 * (1 - a_o1) *  a_h2
        dw7 = (a_o2 - t2) * a_o2 * (1 - a_o2) *  a_h1
        dw8 = (a_o2 - t2) * a_o2 * (1 - a_o2) *  a_h2

        w1 = w1 - (j*dw1)
        w2 = w2 - (j*dw2)
        w3 = w3 - (j*dw3)
        w4 = w4 - (j*dw4)
        w5 = w5 - (j*dw5)
        w6 = w6 - (j*dw6)
        w7 = w7 - (j*dw7)
        w8 = w8 - (j*dw8)
        
    epoch.append(epochlist)
    error.append(Error_total)
    lr.append(j)
  
```

## Plots for change of error wrt epoch for different learning rate

```python

plt.plot(epoch[0], error[0],color ='red', label = lr[0],marker ='o', markersize = 1)
plt.plot(epoch[1], error[1],color ='green', label = lr[1],marker ='o', markersize = 1)
plt.plot(epoch[2], error[2],color ='blue', label = lr[2],marker ='o', markersize = 1)
plt.plot(epoch[3], error[3],color ='orange', label = lr[3],marker ='o', markersize = 1)
plt.plot(epoch[4], error[4],color ='pink', label = lr[4],marker ='o', markersize = 1)
plt.plot(epoch[5], error[5],linestyle ='dashed',color ='black', label = lr[5])

plt.legend()
plt.xlabel('Number of epochs')
plt.ylabel('Total Error')
plt.title('Change in error wrt lr')
plt.show()
         
```
